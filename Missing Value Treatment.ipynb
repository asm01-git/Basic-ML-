```python
#Use KNN to impute null values
gender_df=df
null_index_list=[]
for i in range(len(df)):
    if not pd.isnull(df['gender'][i]):
        gender_df=gender_df.drop(i,axis=0)
    else:
        null_index_list.append(i)
null_index_list
```

```python
from sklearn.model_selection import train_test_split as tts
train_df=df.dropna(axis=0)
train_df=train_df.drop(['customer_id','city','branch_code'],axis=1)
train_df=train_df['gender'].apply(lambda x: 0 if x['gender']=='Male' else 1)
train_df=train_df.astype({'gender':'int64'})
```

```python
x=train_df.drop('gender',axis=1)
y=train_df['gender']
x=pd.get_dummies(x)
train_x,val_x,train_y,val_y=tts(x,y,random_state=40,stratify=y)
train_x.shape,val_x.shape,train_y.shape,val_y.shape
```

```python
#Scale the data
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
train_x_scaled=mms.fit_transform(train_x)
train_x=pd.DataFrame(train_x_scaled,columns=x.columns)
train_x
```

```python
val_x_scaled=mms.fit_transform(val_x)
val_x=pd.DataFrame(val_x_scaled,columns=x.columns)
val_x
```

```python
from sklearn.neighbors import KNeighborsClassifier
k=KNeighborsClassifier(n_neighbors=10)
k.fit(train_x,train_y)
val_pred=k.predict(val_x)
```

```python
from sklearn.metrics import roc_auc_score as ras
ras(val_pred,val_y)
```
